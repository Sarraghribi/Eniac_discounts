```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt


# Creating dataset
discounts_cat = ["aggressive(>50%)","big(25% to 50%)","medium(10% to 25%)","small(<10%)"]

data = [189723.93,1825880.24,5852500.12,6851675.62]


# Creating explode data
explode = (0.1, 0.1, 0.2, 0.3)

# Creating color parameters
colors = ( "firebrick","darkmagenta", "lightsalmon",
		"indianred")

# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "black" }

# Creating autocpt arguments
def func(pct, allvalues):
	absolute = int(pct / 100.*np.sum(allvalues))
	return "{:.1f}%".format(pct, absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
								autopct = lambda pct: func(pct, data),
								explode = explode,
								labels = discounts_cat,
								shadow = True,
								colors = colors,
								startangle = 90,
								wedgeprops = wp,
								textprops = dict(color ="darkblue"))



plt.setp(autotexts, size = 15, weight ="bold")
ax.set_title("discounts_categories in revenue")

# show plot
plt.show()
```
